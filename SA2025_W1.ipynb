### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = df['mpg'].median()
above_median_mpg = df[df['mpg'] > median_mpg]
above_median_mpg['hp_weight_ratio'] = above_median_mpg['horsepower'] / above_median_mpg['weight']
best_ratio_idx = above_median_mpg['hp_weight_ratio'].idxmax()
best_ratio_car = df.loc[best_ratio_idx, 'name']
best_ratio_value = above_median_mpg.loc[best_ratio_idx, 'hp_weight_ratio']

print(f"\nUNGRADED 1: Best HP/weight ratio among above-median MPG cars:")
print(f"Cars: {best_ratio_car}")
print(f"HP/Weight ratio: {best_ratio_value:.4f}")
print(f"MPG: {df.loc[best_ratio_idx, 'mpg']}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
plt.figure(figsize=(12, 6))
avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().unstack()

for origin in avg_mpg_by_year_origin.columns:
    plt.plot(avg_mpg_by_year_origin.index, avg_mpg_by_year_origin[origin],
             marker='o', linewidth=2, label=origin.title())

plt.title('Evolution of Average MPG over years by Origin', fontsize=14, fontweight='bold')
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('average MPG', fontsize=12)
plt.legend(title='Origin')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
print(f"\nUNGRADED 4: Finding 'consistent' car model...")
```

```python
name_year_counts = df.groupby('name')['model_year'].nunique()
multi_year_cars = name_year_counts[name_year_counts > 1].index
```

```python
print(f"Cars appearing in multiple years: {len(multi_year_cars)}")
```

```python
consistent_cars = []
for car_name in multi_year_cars:
    car_data = df[df['name'] == car_name]
    mpg_std = car_data['mpg'].std()
    
    if mpg_std <1.0:
        consistent_cars.append({
            'name': car_name,
            'appearances': len(car_data),
            'avg_mpg': car_data['mpg'].mean(),
            'mpg_std': mpg_std
        })
```

```python
if consistent_cars:
    consistent_df = pd.DataFrame(consistent_cars)
    consistent_df = consistent_df.sort_values(['appearances', 'avg_mpg'], ascending=[False, False])
    
    print(f"\nMost consistent car models (std < 1.0):")
    print(consistent_df.round(3))
    
    print(f"\nMost consistent car: {consistent_df.iloc[0]['name']}")
    print(f"Appearances: {consistent_df.iloc[0]['appearances']}")
    print(f"Average MPG: {consistent_df.iloc[0]['avg_mpg']:.2f}")
    print(f"MPG Standard Deviation: {consistent_df.iloc[0]['mpg_std']:.3f}")
else:
    print("No cars found with standard deviation < 1.0")

print("\n" + "="*50)
print("CONGRATULATIONS! You've completed the Data Grand Prix! ")
print("="*50)
```
